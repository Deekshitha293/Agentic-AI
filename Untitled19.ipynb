```python
def create_calender_event(title, start_time, end_time, attendees, location):
  return f"""
  Event Created Successfully
  Title : {title}
  Start : {start_time}
  End : {end_time}
  Location : {location}
  Attendees : {", ".join(attendees)}
  """
```

```python
def send_email(to, subject, body):
  return f"""
  Email sent successfully
  To : {", ".join(to)}
  Subject : {subject}
  Body : {body}
  """
```

```python
calender_agent = Agent(
    name = "CalanderAgent",
    model = "gemini-2.5-flash",
    description = "Schedules meeting and manages calander.",
    instruction = """
  You are a Calander Assistant.
  Workflow:
  1. First call get_available_slots().
  2. Check whether the requested time is available.
  3. If available, call create_calender_event().
  4. Otherwise suggest another available slot.
  Never create an event before checking availability
  """,
    tools = [create_calender_event, get_available_slots],
)
```

```python
email_agent = Agent(
    name="EmailAgent",
    model="gemini-2.5-flash",
    description="Sends email notifications and confirmations.",
    instruction="""
    You are an Email Assistant.
    Your job is to send email updates or confirmations based on information passed to you from previous steps.
    Always use the send_email tool.
    """,
    tools=[send_email],
)
```

```python
email_agent = Agent(
    name = "EmailAgent",
    model = "gemini-2.5-flash",
    description = "Handles email communication.",
    instruction = """
  You are an Email Assisstant.
  Compose a professional email.
  Include
  Recipient
  Subject
  Body
  Confirmation.
  """,
    tools = [send_email],
)
```

```python
supervisor = Agent(
    name = "SupervisorAgent",
    model = "gemini-2.5-flash",
    description = "Routes tasks to specialized agents.",
    instruction = """
  You are two agents.
  CalenderAgent
  EmailAgent
  If user asks to schedule meeting,
  delegate to CalenderAgent.
  If user asks to send mail,
  delegate to EmailAgent.
  If both are needed,
  delegate to both agents and combine the results.
  """,
    sub_agents = [
        calender_agent,
        email_agent,
    ],
)

runner = InMemoryRunner(supervisor)
```
