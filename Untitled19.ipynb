```python
import os
from google.colab import userdata
os.environ["GOOGLE_GENAI_VERTEXAI"] = "False"
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY1') # Changed GOOGLE_API_KEY1 to GOOGLE_API_KEY
api_key = os.getenv('GOOGLE_API_KEY')
from google.adk.agents import Agent,SequentialAgent,ParallelAgent,LoopAgent
from google.adk.models.google_llm import Gemini
from google.adk.runners import InMemoryRunner
from google.adk.tools import AgentTool,FunctionTool,google_search
from google.genai import types
```

```python
retry_config = types.HttpRetryOptions(
    attempts = 5,
    exp_base = 7,
    initial_delay = 1,
    http_status_codes = [429, 500, 503, 504]
)
```

```python
def create_calender_event(title, start_time, end_time, attendees, location):
  return f"""
  Event Created Successfully
  Title : {title}
  Start : {start_time}
  End : {end_time}
  Location : {location}
  Attendees : {", ".join(attendees)}
  """
```

```python
def send_email(to, subject, body):
  return f"""
  Email sent successfully
  To : {", ".join(to)}
  Subject : {subject}
  Body : {body}
  """
```

```python
os.environ["GOOGLE_API_KEY"] = api_key # Ensure API key is set for agent initialization
calender_agent = Agent(
    name = "CalanderAgent",
    model = "gemini-2.5-flash",
    description = "Schedules meeting and manages calander.",
    instruction = """
  You are a Calander Assistant.
  Workflow:
  1. First call get_available_slots().
  2. Check whether the requested time is available.
  3. If available, call create_calender_event().
  4. Otherwise suggest another available slot.
  Never create an event before checking availability
  """,
    tools = [create_calender_event, get_available_slots],
)
```

```python
os.environ["GOOGLE_API_KEY"] = api_key # Ensure API key is set for agent initialization
email_agent = Agent(
    name = "EmailAgent",
    model = "gemini-2.5-flash",
    description = "Handles email communication.",
    instruction = """
  You are an Email Assisstant.
  Compose a professional email.
  Include
  Recipient
  Subject
  Body
  Confirmation.
  """,
    tools = [send_email],
)
```

```python
supervisor = Agent(
    name = "SupervisorAgent",
    model = "gemini-2.5-flash",
    description = "Routes tasks to specialized agents.",
    instruction = """
  You are two agents.
  CalenderAgent
  EmailAgent
  If user asks to schedule meeting,
  delegate to CalenderAgent.
  If user asks to send mail,
  delegate to EmailAgent.
  If both are needed,
  delegate to both agents and combine the results.
  """,
    sub_agents = [
        calender_agent,
        email_agent,
    ],
)

runner = InMemoryRunner(supervisor)
```

```python
response = await runner.run_debug(
    "Schedule a Zoom meeting with the design team next Tuesday at 2 PM for one hour and send them a reminder email about reviewing the new mockups"
)
```

```python
import asyncio
import nest_asyncio
nest_asyncio.apply()

def main():
  while True:
    question = input(
        "\nask a question (type exit to quit):"
    )

    if question.lower() == "exit":
      print("Exiting...")
      break

    async def run_query(q):
      return await runner.run_debug(q)
    response = asyncio.run(run_query(question))

    print(response)

if __name__ == "__main__":
  main()
```
